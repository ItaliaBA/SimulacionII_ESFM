## Ejercicio 1: 
    a)Usar la función random() para simular 1 volado.
    b)Usar la función random() para simular 1000 volados.
    c)Determinar de b) la probabilidad de que ocurra "sol" en un volado. 

```python
#EJERCICIO 1.a
import random
from collections import Counter

moneda = ['Sol', 'Aguila']
def selectRandom(moneda):
  return random.choice(moneda)

print(selectRandom(moneda))
print("--------------------")

#EJERCICIO 1.b
frecuenciaPalab = []
for i in range(1000):
    mon=selectRandom(moneda)
    print(i+1, mon)
print()
print("--------------------")

#EJERCICIO 1.c
numeros1al2 = range(1,3)
#Sol = 1 y Aguila = 2

def experimento():
    SOL = random.choice(numeros1al2)
    AGUILA = random.choice(numeros1al2)
    return [SOL, AGUILA]


print(experimento())
def suma_es_impar(moneDA):
    return sum(moneDA)%2 == 1

N = 1000
muestra = [experimento() for _ in range(N)]
prob = sum(suma_es_impar(e) for e in muestra)/N
print('Probabilidad aproximada de que salga Sol:', prob)
```
