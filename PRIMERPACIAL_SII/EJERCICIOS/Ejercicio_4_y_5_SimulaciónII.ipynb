## Ejercicio 4:

a) Calcular $E[X_i]$ y $V[X_i]$




Para el experimento de simulación, definimos $Z_N = \frac{X_1 + X_2 + ... + X_N}{N}$

```python
import random
import matplotlib.pyplot as plt 
import numpy as np 
import math
from statistics import mean, variance

def MC(n):
    pi = 0
    lx = []
    ly = []
    counter = 0
    xi = 0
    Z = []
    for i in range(n):
        x = random.random()
        y = random.random()
        if x**2 + y**2 <1:
            counter = counter + 1
            lx.append(counter)
            ly.append(4.0*float(counter)/(i + 1))
            xi = 1
            Z.append(xi)
        else:
            xi = 0
            Z.append(xi)
        pi = 4.0*float(counter)/(i + 1)
        e = np.absolute((pi - np.pi)/np.pi)
        
    p=np.pi/4

    prom = mean(Z)
    errorE = np.absolute((prom - (p))/(p)) 
    var = variance(Z)
    errorV = np.absolute((var - ((p)*(1-p)))/(p))
    print("PI =", pi)
    print("ERROR =", e)
    print("VALOR ESPERADO (SIMULACION)", prom)
    print("ERROR ESPERANZA", errorE)
    print("VARIANZA (SIMULACION)", var)
    print("ERROR VALOR REAL", errorV)
    
    fig = plt.figure(1, figsize=(10, 10))        
    plt.grid()
    plt.plot(lx, ly, color="green", markersize=1)
    plt.show()

MC(80)
```

## Ejercicio 5:
a) Calcular  $E[Z_N]$ y $V[Z_N]$
b) ¿Cuál debe ser el valor de $N$ para que el error sea de 0.01?

Sea $Z_N = \frac{X_1 + X_2 + ... + X_N}{N}$

Para un cuarto de círculo, necesitamos:

$Z_N = 4 \frac{X_1 + X_2 + ... + X_N}{N}$

El valor esperado es:

$E[Z_N]=E[4 \frac{X_1 + X_2 + ... + X_N}{N}] =\frac{4}{N}E[x_{1} + x_{2} + ... +x_{N}] = \frac{4}{N}[E[x_{1}] +  E[x_{2}] + ... + E[x_{N}]] $

$=\frac{4}{N}[NE[x_{i}]]$

$=\frac{4}{N}[N\frac{\pi}{4}]$

$= \pi$

La varianza es:

$V[Z_N]= (\frac{4}{N})^{2}[NV[Z_N]]$

$=\frac{16}{N^{2}}[N(\frac{\pi}{4}-\frac{\pi^{2}}{4^{2}})]$

$ =\frac{16}{N}[\frac{\pi}{4}(1-\frac{\pi}{4})]$

Por lo tanto:

$P(|Z_{N}-\pi|)<\frac{16p(1-p)}{N\epsilon^{2}}$

Si el error es menor que 0.01 $P(|Z_{N}-\pi|)=1$, es decir:

$N < \frac{16p(1-p)}{\epsilon^{2}} = 26967.66$
