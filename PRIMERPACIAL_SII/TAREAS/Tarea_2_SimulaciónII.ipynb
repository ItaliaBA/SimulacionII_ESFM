# **Tarea 2**:

Calcular la integral de raíz cuadrada de arco tangente de 0 a 1, usando un método de MC similar al que usamos para estimar pi.


```python
import random
import matplotlib.pyplot as plt 
import numpy as np 
import math
x = np.linspace(0, 1, 100)
y = np.sqrt(np.arctan(x))
N = 80
m = 0
count = 0 
while count < N:
  x_1 = random.random()
  y_1 = random.random()
  plt.scatter(x_1, y_1, s = 20, c = 'blue')
  z = ((math.atan(x_1)**(1/2))/y_1)
  if z < 1:
    m = m+1
  count = count+1 
fig = plt.figure(1, figsize = (10, 10))
plt.plot(x, y, color = 'green', markersize = 1) 
plt.grid()
plt.show()
pi = 4*(m/N)
print ("RESULTADO: ",pi)
print ("")
```
